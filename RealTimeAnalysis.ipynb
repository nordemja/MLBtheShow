```python
import pandas as pd
import matplotlib.pyplot as plt

hitterDF = pd.read_csv('C:\\Users\\justi\\Documents\\Python Programming\\MLBtheShow\\rankedSeasonsStatsHITTER.csv')
pitcherDF = pd.read_csv('C:\\Users\\justi\\Documents\\Python Programming\\MLBtheShow\\rankedSeasonsStatsPITCHER.csv')

pitcherDF['Date']=pd.to_datetime(pitcherDF['Date'])
pitcherDF
```

```python
totalDf = hitterDF[hitterDF['Player'] == 'Batting Totals']
stadiums = totalDf.groupby('Stadium')['AB', 'R', 'H', 'BB', 'SO'].sum().reset_index().sort_values(by='AB', ascending=False)
stadiums['AVG'] = round(stadiums.H/(stadiums.AB),3)
stadiums['BB%'] = round((stadiums.BB) / (stadiums.AB + stadiums.BB) * 100,2)
stadiums['SO%'] = round((stadiums.SO) / (stadiums.AB + stadiums.BB) * 100, 2)
stadiums = stadiums[stadiums['AB'] > 100]
stadiums = stadiums.sort_values(by='AB', ascending=False,ignore_index=True)
stadiums
```

```python
totalDf = pitcherDF[pitcherDF['Player'] == 'Pitching Totals']
stadiums = totalDf.groupby('Stadium')['IP', 'H', 'R', 'ER', 'SO'].sum().reset_index().sort_values(by='IP', ascending=False)
stadiums['ERA'] = round(9 * (stadiums.ER) / (stadiums.IP) ,2)
#stadiums['SO%'] = round((stadiums.SO) / (stadiums.AB + stadiums.BB) * 100, 2)
stadiums = stadiums[stadiums['IP'] > 25]
stadiums = stadiums.sort_values(by='ERA', ascending=True,ignore_index=True)
stadiums
```

```python
lineup_position = hitterDF.groupby('Lineup')['AB', 'R', 'H', 'BB', 'RBI', 'SO'].sum().reset_index().reset_index()
lineup_position = lineup_position[(lineup_position['Lineup'] < 9) | (lineup_position['Lineup'] == 100) | (lineup_position['Lineup'] == 50)]
lineup_position['AVG'] = round(lineup_position.H / (lineup_position.AB),3)
lineup_position['OBP'] = round((lineup_position.H + lineup_position.BB) / (lineup_position.AB + lineup_position.BB),3)
lineup_position['BB%'] = round((lineup_position.BB) / (lineup_position.AB + lineup_position.BB) * 100,3)
lineup_position['SO%'] = round((lineup_position.SO) / (lineup_position.AB + lineup_position.BB) * 100,3)
print(lineup_position)
```
